# **Import Necessary Libraries**

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager
import numpy as np
import pandas as pd
```

# **Build a basic Graph**

# **Line Chart**

```python
x = np.array([0, 1, 2, 3, 4])
y = np.array([0, 2, 4, 6, 8])

# Resize the graph (dpi specifies pixels per inch. when saving probably should use a higher value if possible)
plt.figure(figsize=(5, 3), dpi=200)

# Line 1

# short hand notation is useful at documentaion 
# format = '[color][marker][line]'
plt.plot(x, y, label='2x', color='blue', linewidth=3, marker='^', linestyle = '--', markersize=7, markeredgecolor = 'blue')

# Line 2
# select interval we want to plot points at
x2 = np.arange(0, 4.5, 0.5)

# plot part of the graph as solid line
plt.plot(x2[:6], x2[:6]**2, 'r', label = 'X^2')

# plot reaminder of the graph as dashed line
plt.plot(x2[5:], x2[5:]**2, 'r--')

# Add title (specify font parameters with fontdict)
plt.title('The first graph', fontdict = {'fontname': 'Adobe Caslon Pro', 'fontsize': 20})

# X and Y Labels
plt.xlabel('X Axis', fontdict = {'fontname': 'Adobe Caslon Pro', 'fontsize': 20})
plt.ylabel('Y Axis', fontdict = {'fontname': 'Adobe Caslon Pro', 'fontsize': 20})

# X and Y axis Tickmarks (scale of your graph )
plt.xticks(x)
plt.yticks(y)

# plt.xticks([0, 1, 2, 3, 4,])
# plt.yticks([0, 2, 4, 6, 8, 10, 12])

# Add a legend
plt.legend()

# Save figure (high dpi is good when saving so graph has higher resolution)
plt.savefig('/content/drive/MyDrive/Deep Learning/Matplolib/my_graph.png', dpi=300)

# Show plot
plt.show()
```

# **Bar Chart**

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]

bars = plt.bar(labels, values)

patterns = ['/', 'O', '*']
for bar in bars:
  bar.set_hatch(patterns.pop(0))

# bars[0].set_hatch('/')
# bars[1].set_hatch('O')
# bars[2].set_hatch('*')

plt.figure(figsize = (6, 4))

plt.show()
```
